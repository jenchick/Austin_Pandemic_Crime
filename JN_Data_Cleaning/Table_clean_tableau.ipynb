```python
import pandas as pd
import numpy as np
import re
```

```python
#2020 data
wide_df = pd.read_csv('Cleaned_final20.csv').drop('Unnamed: 0', 1)
wide_df.head()
```

```python
covid_df = wide_df.iloc[:,:11]
covid_df.head()
```

```python
# use melt function to pivot zipcode to columns and add covid column
covid_melted = pd.melt(covid_df, id_vars = 'zipcode')
covid_melted.head()
```

```python
# fix column labels
covid_melted= covid_melted.rename(columns={'value':'covid cases', 'variable':'date'})
covid_melted.head()
```

```python
# clean date values
covid_melted['date']= covid_melted['date'].str.replace('covid_', '')
covid_melted.head()
```

```python
#crime 2020 data
crime_df=wide_df.iloc[:,np.r_[0, 11:21]]
crime_df.head()
```

```python
# use melt function to pivot zipcode to columns and add covid column

crime_melted = pd.melt(crime_df, id_vars = 'zipcode')
crime_melted.head()
```

```python
# fix column labels

crime_melted= crime_melted.rename(columns={'value':'crime cases', 'variable':'date'})
crime_melted.head()
```

```python
# clean date values

crime_melted['date']= crime_melted['date'].str.replace('crime_', '')
crime_melted.head()
```
