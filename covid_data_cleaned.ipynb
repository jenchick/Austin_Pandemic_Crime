```python
# import Dependencies 
import pandas as pd
import numpy as np
```

```python
# read in csv file
covid_df = pd.read_csv("covid_zip.csv", parse_dates=['Date'], dayfirst= True)
covid_df.head()
```

```python
# removing "ZIP_ from column name"
covid_df.columns = covid_df.columns.str.replace("Zip_", "")
covid_df.head()
```

```python
# drop notes column and FID columns
covid_df = covid_df.drop(columns=['FID', 'Notes'])
covid_df.head()
```

```python
# Convert datetime to date
covid_df['Date'] = covid_df['Date'].dt.strftime('%m/%d/%Y')
covid_df.head()
```

```python
# export to csv
covid_df.to_csv('covid_cases_zip.csv')
```
