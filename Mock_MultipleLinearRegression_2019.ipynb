```python
# Import dependencies 
import pandas as pd 
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import r2_score
import numpy as np
import datetime
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Import mock covid zipcodes csv
file_path = 'mock_covid_zipcodes.csv'
mock_zipcodes_df = pd.read_csv(file_path)
mock_zipcodes_df.head(10)
```

```python
# Import mock original crimedata csv
file_path = 'mock_original_crimedata.csv'
mock_crimedata_df = pd.read_csv(file_path)
mock_crimedata_df.head(10)
```

```python
# Rename column hh to aa
mock_zipcodes_df.rename(columns = {'hh':'aa'}, inplace = True)
mock_zipcodes_df.head(10)
```

```python
# Merge dataframes
mock_df = pd.concat([mock_zipcodes_df.set_index('aa'),mock_crimedata_df.set_index('aa')], axis=1, join='inner')
mock_df.head(10)
```

```python
# Drop duplicates
mock_df=mock_df.T.drop_duplicates().T
mock_df.head(10)
```

```python
# Convert date column
mock_df['dates'] = pd.to_datetime(mock_df['dates'])
mock_df.head(10)
```

```python
# Drop days
mock_df['date'] = pd.to_datetime(mock_df['dates']).dt.to_period('M')
```

```python
# Print dataframe
mock_df.head(10)
```

```python
# Drop columns 
mock_df = mock_df.drop(columns=['dates'], axis=1)
```

```python
# Print mock_df
mock_df.head(10)
```

```python
# Use label encoder to convert column "ii"
le = LabelEncoder()
mock_df2 = mock_df.copy()
mock_df2['ii'] = le.fit_transform(mock_df2['ii'])
```

```python
# Print new dataframe
mock_df2.head(10)
```

```python
# Set x and y
X = mock_df2 [['ii', 'gg']]
y = mock_df2[['bb']]
```

```python
# Split into train and test sets, with 70, 30 split
X_train, X_test, y_train, y_test = train_test_split (X,y, random_state = 75, train_size=0.70)
```

```python
# Print shapes
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Create model
precovid_model = LinearRegression()
```

```python
# Fit model
precovid_model.fit(X,y)
```

```python
# Predict the model
pred = precovid_model.predict(X_test)
print(pred.shape)
```

```python
# Print model parameters
print(precovid_model.coef_)
print(precovid_model.intercept_)
```

```python
r_squared = r2_score(y_test, pred)
print(r_squared)
```
